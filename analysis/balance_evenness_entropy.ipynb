```python
def gen_complex_sine(k, N):
    """Generate one of the complex sinusoids used in the DFT from its frequency index and the DFT length.
    
    Args:
        k (integer): frequency index of the complex sinusoid of the DFT
        N (integer) = length of complex sinusoid, DFT length, in samples
        
    Returns:
        np.array: array with generated complex sinusoid (length N)
        
    """
    ### your code here
    n = np.arange(N)

    return np.exp((-1) * 1j * 2 * np.pi * k * n / N)
def dft(x):
    """Compute the DFT of a signal.
    
    Args:
        x (numpy array): input sequence of length N
        
    Returns:
        np.array: N point DFT of the input sequence x
    """
    ## Your code here
    X = np.array([])
    n = np.arange(x.size)
    for k in range(x.size):
      complSine = gen_complex_sine(k,x.size)
      X = np.append(X,sum(x*(complSine)))
    return X
def gen_mag_spec(x):
    """Compute magnitude spectrum of a signal.
    
    Args:
        x (np.array): input sequence of length N
        
    Returns:
        np.array: magnitude spectrum of the input sequence x (length N)
        
    """
    ### your code here
    X = np.array([])
    n = np.arange(x.size)
    for k in range(x.size):
      complSine = gen_complex_sine(k,x.size)
      X = np.append(X, sum(x * complSine))
    return np.abs(X)

def balance(DFT:np.array):
    balance = []
    for i in DFT:
       balance.append( 1 - (i)/DFT.size ) 
    return balance
```

```python

#testing balance
#generate signal
x = np.cos(2*np.pi*200.0*np.arange(512)/1000)
#getting dft
signal_dft = dft(x)
#magnitude spectrum
mag_spectrum = gen_mag_spec(signal_dft)
#balance
signal_balance = balance(mag_spectrum)
print(signal_balance)
print(sum(signal_balance)/len(signal_balance))
```
