## Drum Mappings for RhythmToolbox

GM_dict={
 key is midi note number
 values are:
 
 [0] name (as string)
 
 [1] name category low mid or high (as string)
 
 [2] substiture midi number for simplified MIDI (all instruments)
 
 [3] name of instrument for 8 note conversion (as string)
 
 [4] number of instrument for 8 note conversion 
 
 [5] substiture midi number for conversion to 8 note
 
 [6] substiture midi number for conversion to 16 note
 
 [7] substiture midi number for conversion to 3 note
 



```python
from hvo_sequence.test.feature_masking_experiment.rhythmtoolbox.rhythmtoolbox.midi_mapping import *
```

```python
_8note_map = {}
for midi_val, data in GM_dict.items():
    inst_name = GM_dict[data[5]][0]
    inst_num = data[5]
    if inst_name not in _8note_map.keys():
        _8note_map.update({inst_name: [inst_num]})
    _8note_map[inst_name].append(midi_val)
    
_16note_map = {}
for midi_val, data in GM_dict.items():
    inst_name = GM_dict[data[6]][0]
    inst_num = data[6]
    if inst_name not in _16note_map.keys():
        _16note_map.update({inst_name: [inst_num]})
    _16note_map[inst_name].append(midi_val)
    
_3note_map = {}
for midi_val, data in GM_dict.items():
    inst_name = GM_dict[data[7]][0]
    inst_num = data[7]
    if inst_name not in _3note_map.keys():
        _3note_map.update({inst_name: [inst_num]})
    _3note_map[inst_name].append(midi_val)

    
```

```python
from data import load_gmd_hvo_sequences

train_set = load_gmd_hvo_sequences(
    dataset_setting_json_path = "data/dataset_json_settings/4_4_Beats_gmd.json", 
    subset_tag = "train", 
    force_regenerate=False)

_1bar_set = []

# load with 16 steps only
for ix, exp_hvo_seq in enumerate(train_set):
    temp = exp_hvo_seq.copy_empty()
    if (exp_hvo_seq.hvo[16:, :].shape[0]>0):
        temp.hvo = exp_hvo_seq.hvo[16:, :]
        # temp.offsets = np.zeros(temp.offsets.shape)
        _1bar_set.append(temp)
    
```

```python
from hvo_sequence.test.feature_masking_experiment.rhythmtoolbox import rhythmtoolbox
```

```python
sample_ix = 10000
_1bar_set[sample_ix].to_html_plot(show_figure=True)
patlist = _1bar_set[sample_ix].get_per_step_list_of_active_voices(use_label_instead_of_midi_number=False, print_mapping=False)
print(rhythmtoolbox.pattlist2descriptors(patlist))
Audio(data=_1bar_set[sample_ix].synthesize(sf_path="hvo_sequence/soundfonts/Standard_Drum_Kit.sf2"), rate=44100)
```

```python
masked_version = _1bar_set[sample_ix].get_masked_variation(ratio_of_hits_to_mask = 0.3,
                                                           mask_creation_mode=1)
masked_version.to_html_plot(show_figure=True)
masked_patlist = masked_version.get_per_step_list_of_active_voices(use_label_instead_of_midi_number=False, print_mapping=False)
print(rhythmtoolbox.pattlist2descriptors(masked_patlist))
Audio(data=masked_version.synthesize(sf_path="hvo_sequence/soundfonts/Standard_Drum_Kit.sf2"), rate=44100)
```
